```python
from EulerMaruyama import EulerMaruyama
import numpy as np
```

```python
def mu(x,t):
    return x[1],-9.8

def sigma(x,t):
    return 0.0,1.0
```

```python
em = EulerMaruyama(1000,0,10,100,[0,47.5,],mu=mu,sigma=sigma)
print('Velocity:')
em.plot_solution(eq=1,title='Velocity Y')
print('Position:')
em.plot_solution(eq=0,title='Position Y')
```

```python
em.plot_dist(10,0)
```

```python
import matplotlib.pyplot as plt
tempy,tempx=[],[]
for ys in em.solution:
    tempy.append(ys[-1,1])
    tempx.append(ys[-1,0])
plt.scatter(tempx,tempy)
plt.xlabel('Position of Particle in Y')
plt.ylabel('Velocity of Particle in Y')
plt.show()
```
